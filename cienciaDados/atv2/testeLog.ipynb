```python
# MONITOR — rode quando quiser ver o progresso
import json
from pathlib import Path

CACHE_DIR = Path("cache_experimentos")

if not CACHE_DIR.exists():
    print("Pasta cache_experimentos/ nao encontrada.")
else:
    prog_files = sorted(CACHE_DIR.glob("progress_*.json"))
    done_files = sorted(CACHE_DIR.glob("cache_*.json"))

    print(f"Jobs completos : {len(done_files)}/12")
    for f in done_files:
        print(f"  [DONE] {f.stem}")

    print(f"\nJobs em andamento: {len(prog_files)}")
    for pf in prog_files:
        with open(pf) as f:
            p = json.load(f)
        n = len(p.get("exp_cache", {}))
        pct = n / 21 * 100
        bar = "#" * int(pct / 5) + "-" * (20 - int(pct / 5))
        print(f"  [{bar}] {p['key']} seed={p['seed']}: {n}/21 ({pct:.0f}%)")

    total_done = len(done_files) * 21 + sum(
        len(json.load(open(pf)).get("exp_cache", {})) for pf in prog_files
    )
    total = 12 * 21
    print(f"\nProgresso geral: {total_done}/{total} experimentos ({total_done/total*100:.1f}%)")
```
