# Utilitários — Experimento Lab 2

Este notebook contém ferramentas auxiliares para gerenciar o cache e monitorar o progresso da execução.


```python
# ─────────────────────────────────────────────────────────────
# MONITOR — rode a qualquer momento para ver o progresso
# Nao interfere na execucao principal
# ─────────────────────────────────────────────────────────────

import json
from pathlib import Path

CACHE_DIR = Path("cache_experimentos")

if not CACHE_DIR.exists():
    print("Pasta cache_experimentos/ nao encontrada.")
    print("Verifique se este notebook esta na mesma pasta que o lab2Test.ipynb")
else:
    prog_files = sorted(CACHE_DIR.glob("progress_*.json"))
    done_files = sorted(CACHE_DIR.glob("cache_*.json"))

    print(f"Jobs completos : {len(done_files)}/12")
    for f in done_files:
        print(f"  [DONE] {f.stem}")

    print(f"\nJobs em andamento: {len(prog_files)}")
    for pf in prog_files:
        with open(pf) as f:
            p = json.load(f)
        n = len(p.get("exp_cache", {}))
        pct = n / 21 * 100
        bar = "#" * int(pct / 5) + "-" * (20 - int(pct / 5))
        print(f"  [{bar}] {p['key']} seed={p['seed']}: {n}/21 ({pct:.0f}%)")

    total_done = len(done_files) * 21 + sum(
        len(json.load(open(pf)).get("exp_cache", {})) for pf in prog_files
    )
    total = 12 * 21
    print(f"\nProgresso geral: {total_done}/{total} experimentos ({total_done/total*100:.1f}%)")
```

## Limpeza de cache

⚠️ **Atenção:** rode apenas quando quiser **resetar** todos os resultados. Isso apaga o progresso já calculado.

```python
# ─────────────────────────────────────────────────────────────
# LIMPEZA — apaga todo o cache e reseta o progresso
# Execute manualmente apenas quando quiser refazer tudo do zero
# ─────────────────────────────────────────────────────────────

import glob, os
from pathlib import Path

CACHE_DIR = Path("cache_experimentos")
CACHE_DIR.mkdir(exist_ok=True)

removidos = 0

# cache antigo em data/ (execucoes anteriores)
for f in glob.glob("data/cache_*.json") + glob.glob("data/progress_*.json"):
    os.remove(f)
    print(f"Removido (data/): {f}")
    removidos += 1

# cache na pasta dedicada
for f in CACHE_DIR.glob("*.json"):
    os.remove(f)
    print(f"Removido (cache_experimentos/): {f}")
    removidos += 1

if removidos == 0:
    print("Nenhum cache encontrado. Tudo limpo.")
else:
    print(f"\n{removidos} arquivo(s) removido(s).")

print("\nPasta cache_experimentos/ pronta. Pode rodar a celula de execucao.")
```
